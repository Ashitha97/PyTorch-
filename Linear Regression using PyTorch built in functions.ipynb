```python
import torch.nn as nn
import torch
import numpy as np
```

## As i have done a linear regression using pytorch without built in methods was a quite tedious way, we can build a model using simple built in methods in PyTorch

### let us take inputs and targets with larger size , we are using a 15 training examples this time

```python
inputs=np.array([[73,67,43],[91,88,64],[87,134,58],
                [102,43,37],[69,96,70],[45,78,102],
                [35,104,65],[105,99,70],[102,30,45],
                [67,89,100],[89,76,56],[65,37,23],
                [98,89,66],[102,10,90],[78,65,22]],dtype='float32')


targets=np.array([[56,70],[89,45],[78,90],
                  [56,43],[89,10],[119,133],
                  [103,119],[81,101],[22,37],
                  [117,80],[90,78],[21,22],
                  [34,56],[103,110],[23,45]],dtype='float32')

inputs=torch.from_numpy(inputs)
targets=torch.from_numpy(targets)
```

```python
targets.shape
```

# Dataset and Dataloader

```python
from torch.utils.data import TensorDataset,DataLoader
```

```python
#define dataset
#tensordataset is a special kind of tensor dataset
#tensor dataset allows us to acces small section of our training data using the indexing
train_ds=TensorDataset(inputs,targets)
train_ds[0:3]
```

- we will also create dataloader, which can split the data into batches of predefined size while training it aslo provides shuffling and random sampling

- usually we work in batches as we have a very larget datasets suppose say millions we can make use of this batch_size in data loader



```python
#define data loader
batch_size=5
train_dl=DataLoader(train_ds,batch_size,shuffle=True)
```

- DataLoader is typically used in for-in loop

```python
for xb,yb in train_dl:
    print(xb)
    print(yb)
    break
```

- instead of manually defining weights all the time we can define a model using nn.linear, which does it automatically

```python
## we can define model using nn.linear
model=nn.Linear(3,2)# 3 is number of inputs and 2 is number of outputs
print(model.weight)
print(model.bias)
```

```python
#parameters
list(model.parameters())
```

```python
#generate predictions
preds=model(inputs)
```

preds

#pretty bad predictions ,because the model is randomly initialized

## Loss function

- insetad of defining a loss function manually we can use a built in loss fuction

```python
##importing nnfunctional
import torch.nn.functional as F
```

```python
loss_fn=F.mse_loss
```

```python
loss=loss_fn(model(inputs),targets)
```

```python
print(loss)
```

## Optimizer

- instead of manually manipulating weights and biases we can use the optimizer,optim.SGD that is stochastic gradienr descent,it is stochastic because samples were selected in batches instead of taking the entire training data

```python
#define optimizer
opt=torch.optim.SGD(model.parameters(),lr=1e-5)
```

```python
##utility functions to train the model
def fit(num_epochs,model,loss_fn,opt,train_dl):
    #repeat for given number of epochs
    for epoch in range(num_epochs):
        
        #train with batches of data
        for xb,yb in train_dl:
            
            #generate predictions
            pred=model(xb)
            
            #calculate loss
            loss=loss_fn(pred,yb)
            
            #compute gradients
            loss.backward()
            
            #update parameters using gradients
            opt.step()
            
            #reset the gradients back to zero
            opt.zero_grad()
            
            
            #print the progress
            if (epoch+1) % 10==0:
                     print('Epoch [{}/{}], Loss: {:.4f}'.format(epoch+1, num_epochs, loss.item()))
```

```python
fit(100000,model,loss_fn,opt,train_dl)
```
